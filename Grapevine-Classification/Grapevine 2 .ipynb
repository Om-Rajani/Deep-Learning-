```python
import os 
import seaborn as sns 
import glob
import cv2
import numpy as np 
import tensorflow.keras.utils as ut 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

# LOADING THE DATASET

```python
print(os.listdir("D:\Artificial intelligence\Grapevine Leaves Image\Grapevine_Leaves_Image_Dataset/"))
```

```python
size = 224
```

```python
images=[]
labels = []

for dir_path in glob.glob('D:\Artificial intelligence\Grapevine Leaves Image\Grapevine_Leaves_Image_Dataset/*'):
    label = dir_path.split('\\')[-1]
    print(label)
    for img_path in glob.glob(os.path.join(dir_path,"*.png")):
        print(img_path)
        img = cv2.imread(img_path,cv2.IMREAD_COLOR)
        img = cv2.resize(img,(size,size))
        img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
        images.append(img)
        labels.append(label)

images = np.array(images)
labels = np.array(labels)
```

```python
from keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator( 
        zca_whitening=False,  
        rotation_range=10,  
        zoom_range = 0.1,  
        width_shift_range=0.1,  
        height_shift_range=0.1, 
        horizontal_flip=True,  
        vertical_flip=True)  
```

```python
train_generator = datagen.flow_from_directory(
    directory=r"D:\Artificial intelligence\Grapevine Leaves Image\Grapevine_Leaves_Image_Dataset/",
    target_size=(224, 224),
    color_mode="rgb",
    batch_size=32,
    class_mode="categorical",
    shuffle=True,
    seed=42
)
```

```python
le = LabelEncoder()
le.fit(labels)
labels_encoded = le.transform(labels)
```

```python
x_train, x_validate, y_train, y_validate = train_test_split(images,labels_encoded, test_size = 0.1, random_state = 2)
```

```python
x_train, x_validate = x_train / 255.0, x_validate / 255.0
```
